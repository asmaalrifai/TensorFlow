# Sarima

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plot style
sns.set(style="whitegrid")

results_dir = os.path.join(os.getcwd(), "results")
graphs_dir = os.path.join(results_dir, "graphs")
```

**1. Display SARIMA Model Results**

```python
# Load SARIMA results and averages
sarima_results_df = pd.read_csv(r"../results/sarima_results.csv")
sarima_avg_results_df = pd.read_csv(r"../results/sarima_results_avg.csv")

# Display SARIMA Results
print("SARIMA Model Results:")
display(sarima_results_df)

print("\nAverage SARIMA Results:")
display(sarima_avg_results_df)
```

**2. Plot Actual vs Predicted for SARIMA Models**

```python
# Define the test files and prediction files
test_files = ['test_data_1', 'test_data_2', 'test_data_3']
prediction_files = [
    os.path.join(results_dir, "sarima_predictions_test_data_1.csv"),
    os.path.join(results_dir, "sarima_predictions_test_data_2.csv"),
    os.path.join(results_dir, "sarima_predictions_test_data_3.csv")
]

# Plot the actual vs predicted for each of the test data
for test_file, prediction_file in zip(test_files, prediction_files):
    # Load the predictions
    predictions_df = pd.read_csv(prediction_file, parse_dates=["Tarih"], index_col="Tarih")
    
    # Plotting Actual vs Predicted for each test file
    plt.figure(figsize=(12, 6))
    plt.plot(predictions_df.index, predictions_df["Actual"], label="Actual", color="blue", marker="o")
    plt.plot(predictions_df.index, predictions_df["Predicted"], label="Predicted (SARIMA)", color="red", linestyle="dashed", marker="s")
    
    plt.title(f"SARIMA Model: Actual vs Predicted for {test_file}", fontsize=16)
    plt.xlabel("Date")
    plt.ylabel("Ptfdolar")
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    
    # Save the plot
    plt.savefig(f"../results/graphs/sarima_{test_file}.png")
    plt.show()

print("SARIMA model results and graphs have been successfully displayed and saved.")
```

# ML Models (SVR, XGBoost)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set plot style
sns.set(style="whitegrid")

# Define relative paths for ML results and averages
results_dir = os.path.join(os.getcwd(), "../results")
graphs_dir = os.path.join(results_dir, "graphs")
```

```python


# Ensure results directories exist
os.makedirs(graphs_dir, exist_ok=True)

# Load ML results and averages using relative paths
ml_results_df = pd.read_csv(r"../results/ml_results.csv")
ml_avg_results_df = pd.read_csv(r"../results/ml_results_avg.csv")

# Display ML Results
print("Machine Learning Model Results (SVR & XGBoost):")
display(ml_results_df)

print("\nAverage Machine Learning Model Results:")
display(ml_avg_results_df)
```

Compare Model Performance

```python
# ------------------------------
# Comparing the MAE, SMAPE, and RMSE between SVR and XGBoost Models
# ------------------------------

# Plot MAE comparison for SVR and XGBoost
plt.figure(figsize=(12, 6))
sns.barplot(x="Test File", y="MAE", hue="Model", data=ml_results_df, palette="muted")
plt.title("Comparison of MAE between SVR and XGBoost Models", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot SMAPE comparison for SVR and XGBoost
plt.figure(figsize=(12, 6))
sns.barplot(x="Test File", y="SMAPE (%)", hue="Model", data=ml_results_df, palette="muted")
plt.title("Comparison of SMAPE between SVR and XGBoost Models", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot RMSE comparison for SVR and XGBoost
plt.figure(figsize=(12, 6))
sns.barplot(x="Test File", y="RMSE", hue="Model", data=ml_results_df, palette="muted")
plt.title("Comparison of RMSE between SVR and XGBoost Models", fontsize=16)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 Display Actual vs Predicted Graphs for SVR and XGBoost Models

```python
# ------------------------------
# Displaying and comparing graphs for each test data file (SVR & XGBoost)
# ------------------------------

# Define the paths for the prediction files
test_files = ['test_data_1', 'test_data_2', 'test_data_3']
svr_files = [
    os.path.join(results_dir, "svr_predictions_test_data_1.csv"),
    os.path.join(results_dir, "svr_predictions_test_data_2.csv"),
    os.path.join(results_dir, "svr_predictions_test_data_3.csv")
]
xgb_files = [
    os.path.join(results_dir, "xgb_predictions_test_data_1.csv"),
    os.path.join(results_dir, "xgb_predictions_test_data_2.csv"),
    os.path.join(results_dir, "xgb_predictions_test_data_3.csv")
]

# Plot the actual vs predicted for each of the test data (SVR and XGBoost)
for test_file, svr_file, xgb_file in zip(test_files, svr_files, xgb_files):
    # Load the SVR predictions
    svr_predictions_df = pd.read_csv(svr_file, parse_dates=["Tarih"], index_col="Tarih")
    # Load the XGBoost predictions
    xgb_predictions_df = pd.read_csv(xgb_file, parse_dates=["Tarih"], index_col="Tarih")

    # Plotting Actual vs Predicted for SVR
    plt.figure(figsize=(12, 6))
    plt.plot(svr_predictions_df.index, svr_predictions_df["Actual"], label="Actual", color="blue", marker="o")
    plt.plot(svr_predictions_df.index, svr_predictions_df["Predicted"], label="Predicted (SVR)", color="red", linestyle="dashed", marker="s")
    plt.title(f"SVR Model: Actual vs Predicted for {test_file}", fontsize=16)
    plt.xlabel("Date")
    plt.ylabel("Ptfdolar")
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

    # Plotting Actual vs Predicted for XGBoost
    plt.figure(figsize=(12, 6))
    plt.plot(xgb_predictions_df.index, xgb_predictions_df["Actual"], label="Actual", color="blue", marker="o")
    plt.plot(xgb_predictions_df.index, xgb_predictions_df["Predicted"], label="Predicted (XGBoost)", color="green", linestyle="dashed", marker="s")
    plt.title(f"XGBoost Model: Actual vs Predicted for {test_file}", fontsize=16)
    plt.xlabel("Date")
    plt.ylabel("Ptfdolar")
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```
